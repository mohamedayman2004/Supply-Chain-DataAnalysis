## Feature Engineering

```python
# Profit Calculation
data['Profit'] = data['Revenue generated'] - (
   data['Manufacturing costs'] + data['Shipping costs'] + data['Costs']
)
```

```python
# Total Lead Time
data['Total Lead Time'] = data['Lead time'] + data['Manufacturing lead time'] + data['Shipping times']
```

```python
## Average Order Value
data['Average Order Value'] = data['Revenue generated'] / data['Number of products sold']
```

```python
# High Defect Indicator
data['High Defect'] = data['Defect rates'] > 3  
```

## Exploratory Data Analysis (EDA)

#### 1) Product Performance

```python
# Revenue by product type
data.groupby('Product type')['Revenue generated'].sum().sort_values(ascending=False)

# Profit by SKU
data.groupby('SKU')['Profit'].sum().sort_values(ascending=False)
```

## Visualizations

#### 📊 Revenue by Product Type

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Product type', y='Revenue generated', estimator=sum)
plt.title("Total Revenue by Product Type", fontsize=16)
plt.xlabel("Product Type")
plt.ylabel("Total Revenue")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

#### ⏱️ Total Lead Time by Supplier

```python
plt.figure(figsize=(12, 6))
top_suppliers = data['Supplier name'].value_counts().nlargest(10).index
sns.barplot(data=data[data['Supplier name'].isin(top_suppliers)], 
            x='Supplier name', y='Total Lead Time', ci=None)
plt.title("Average Total Lead Time by Top 10 Suppliers", fontsize=16)
plt.xlabel("Supplier Name")
plt.ylabel("Total Lead Time (days)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 🧩 Supplier Share (Pie Chart)

```python
plt.figure(figsize=(8, 8))
data['Supplier name'].value_counts().nlargest(5).plot.pie(autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Supplier Share", fontsize=16)
plt.ylabel("")  # Hides default 'Supplier name'
plt.tight_layout()
plt.show()
```

#### 📦 Shipping Cost Distribution by Transportation Mode

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Transportation modes', y='Shipping costs')
plt.title("Shipping Cost Distribution by Transportation Mode", fontsize=16)
plt.xlabel("Transportation Mode")
plt.ylabel("Shipping Cost")
plt.tight_layout()
plt.show()
```

####  🔥 Correlation Heatmap (Numerical Insights)

```python
plt.figure(figsize=(12, 8))
numeric_corr = data.corr(numeric_only=True)
sns.heatmap(numeric_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

## Clustering Suppliers (KMeans)

```python
from sklearn.preprocessing import StandardScaler

# Select relevant features
supplier_features = data.groupby('Supplier name')[['Total Lead Time', 'Costs']].mean()

# Scale features
scaler = StandardScaler()
supplier_scaled = scaler.fit_transform(supplier_features)
```

```python
from sklearn.cluster import KMeans

# Choose number of clusters (try 3–5)
kmeans = KMeans(n_clusters=4, random_state=42)
supplier_features['Cluster'] = kmeans.fit_predict(supplier_scaled)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=supplier_features,
    x='Total Lead Time', y='Costs',
    hue='Cluster', palette='Set2', s=100
)
plt.title("Supplier Clustering by Lead Time and Cost")
plt.xlabel("Average Total Lead Time")
plt.ylabel("Average Cost")
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```
